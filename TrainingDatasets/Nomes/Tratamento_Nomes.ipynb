<h1> Tratamentamento de nomes brasileiros </h1>


Esse notebook tem como objetivo tratar o banco de dados de nome retirado do site:

https://brasil.io/dataset/genero-nomes/nomes/|

```python
# importando bibliotecas
import pandas as pd
import numpy as np
```

```python
# importando dataset
df = pd.read_csv("nomes_brasil.csv")
```

```python
#olhando o data_set
df
```

```python
print(df.columns.values)
```

```python
df = df[df["frequency_total"]>8000]
df
```

```python
#elimindando colunas necessárias
df1 = df.drop(["name", "classification", "frequency_female", "frequency_male", "frequency_total", "ratio"], axis=1)
```

```python
df1
```

```python
#salvando df1
df1.to_csv("nomes_pre_tratados_famosos.csv", index=False)
```

```python
#iterando no meu df e criando uma lista com cada linha 
names = [x for x in df1["names"]]
```

```python
#tratar a lista tirando as barras
nomes_tratados = []
for x in names:
    for y in x.split("|"):
        nomes_tratados.append(y)    
```

```python
nomes_tratados
```

```python
def remove_values_from_list(the_list, val):
   return [value for value in the_list if value != val]
```

```python
nomes_tratados = remove_values_from_list(nomes_tratados, '')
```

```python
nomes_tratados
```

```python
string_list = [each_string.lower() for each_string in nomes_tratados]
```
